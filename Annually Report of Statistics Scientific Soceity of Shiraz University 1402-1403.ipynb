```python
dataframes = [df1, df2, df3, df4, df5, df6, df7, df8, df9]

# 1. Get the union of all column names across the DataFrames
all_columns = set().union(*[df.columns for df in dataframes])
reindexed_dfs = [df.reindex(columns=all_columns) for df in dataframes]
combined_df = pd.concat(reindexed_dfs, axis=0, ignore_index=True)
```

```python
# Assuming you have the combined_df DataFrame already available
category_counts = combined_df['جنسیت'].value_counts(dropna=True)

# Define a modern color palette
colors = ['#66c2a5', '#fc8d62', '#8da0cb', '#e78ac3', '#a6d854']

# Create the pie chart with enhanced style
plt.figure(figsize=(8, 8))  # Set figure size for better visualization
plt.pie(
    category_counts,
    labels=category_counts.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=colors,
    shadow=True,  # Add shadow for a 3D effect
    wedgeprops={'edgecolor': 'white', 'linewidth': 2},  # Modern edge styling
    textprops={'fontsize': 14, 'color': 'black'}  # Modern font style
)

# Add a title with a modern font
plt.title('tozie daneshjoyan dorehaye amoozeshi bar asas jensiat anha', fontsize=16, fontweight='bold', color='#333333')

# Display the plot
plt.show()
```

```python
from wordcloud_fa import WordCloudFa
```

```python
from PIL import Image
iran_mask = np.array(Image.open('/content/N82562558-71674576.jpg'))
```

```python
#Assuming you already have the frequencies dictionary
frequencies = daneshgah_mahal_thasil.value_counts().to_dict()

# Function to generate darker colors based on frequency
def color_func(word, font_size, position, orientation, random_state=None, **kwargs):
    # Generate a color from a darker color palette based on the font size (frequency)
    # Here, we use a darker shade of blue, green, and red
    # You can customize this further with different colors
    return f'rgba(0, 0, {int(255 * (font_size / 200))}, 1)'  # Dark blueish based on font size

# Initialize the Persian WordCloud
wordcloud = WordCloudFa(
    persian_normalize=True,
    no_reshape=True,
    width=800,
    height=400,
    font_path=font_path,
    mask = iran_mask,
    background_color="white",
    min_font_size=8,
    contour_color='black',
    contour_width=4,
    max_font_size=200,
    max_words=300,
    color_func=color_func  # Use the custom color function
)

# Generate word cloud from frequencies
wc = wordcloud.generate_from_frequencies(frequencies)

# Display the word cloud using matplotlib
plt.figure(figsize=(10, 8))
plt.imshow(wc, interpolation='bilinear')  # Use bilinear interpolation for smooth rendering
plt.axis('off')  # No axes around the word cloud

# Title with modern styling
plt.title(
    "namayeshi az faravani daneshgahaye mahal tahsil daneshjooyan dore haye amoozeshi be soorat abr kalamat",
    fontsize=24,  # Increase title font size
    fontweight='bold',  # Bold title for emphasis
    color='#333333',  # Dark gray color for title
    loc='center',  # Center the title
    pad=20  # Padding for title spacing
)

plt.show()
```
